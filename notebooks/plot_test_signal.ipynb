## Imports

```python
import numpy as np
import h5py
from matplotlib import pyplot as plt
```

## Load test signal from file

```python
input_fpath = '/Users/cmcgrory/park_lab/realtime/basic_prototype/data/processed/r11_avg_fr.h5'

with h5py.File(input_fpath, 'r') as f:
    fr_avg = f['fr_avg_hz'][:]
```

## Plot part of signal

```python
# TODO: Load this from file
bin_size = 0.01

# Number of data points to plot
tmax_plot = 300

# Plot first part of signal
tvals_plot = bin_size * np.arange(tmax_plot)
fig, ax = plt.subplots()
plt.plot(tvals_plot, fr_avg[0:tmax_plot])
ax.set(xlabel='time (s)', ylabel='avg. firing rate (Hz)');
```
